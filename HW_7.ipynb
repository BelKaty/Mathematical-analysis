## 1. Найти интервалы возрастания и убывания функций:

1) $ f(x) = x + e^{-x} $

Производная функции.

$ f'(x) = (x + e^{-x})' = 1 - e^{-x} $

Определим, где производная положительна, где отрицательна и где равна нулю.

$ f'(x) > 0 $ при $ 1 - e^{-x} > 0 $

$ e^{-x} < 1 $

$ -x < 0 $

$ x > 0 $

$ f'(x) < 0 $ при $ 1 - e^{-x} < 0 $

$ e^{-x} > 1 $

$ -x > 0 $

$ x < 0 $

$ f'(x) = 0 $ при $ 1 - e^{-x} = 0 $

$ e^{-x} = 1 $

$ -x = 0 $

$ x = 0$

Значит:

- функция возрастает на интервале $ (0, +\infty) $.
- функция убывает на интервале $ (-\infty, 0) $.
- функция имеет минимум в точке $ x = 0$

2) $ f(x) = x \ln x $

Найдем производную функции.

$ f'(x) = (x \ln x)' = \ln x + 1 $

Определим интервалы знакопостоянства производной.

$ f'(x) > 0 $ при $ \ln x + 1 > 0 $

$ \ln x > -1 $

$ x > e^{-1} $

$ x > \frac{1}{e} $

$ f'(x) < 0 $ при $ \ln x + 1 < 0 $

$ \ln x < -1 $

$ x < e^{-1} $

$ x < \frac{1}{e} $

Найдем нули функции.

$ f'(x) = 0 $ при $ \ln x + 1 = 0 $

$ \ln x => -1 $

$ x = e^{-1} $

$ x = \frac{1}{e} $

Получаем:

- функция возрастает на интервале $ \left( \frac{1}{e}, +\infty \right) $.
- функция убывает на интервале $ (0, \frac{1}{e}) $.
- функция имеет минимум в точке $ x = \frac{1}{e} $

## 3. Найти интервалы выпуклости и точки перегиба функций:

1) $ f(x) = e^{-x^2} $

Найдем первую производную функции.

$ f'(x) = (e^{-x^2})' = e^{-x^2} \cdot (-2x) = -2x e^{-x^2} $

Найдем вторую производную функции.

$ f''(x) = (-2x e^{-x^2})' = -2 e^{-x^2} + (-2x) \cdot (-2x e^{-x^2}) = -2 e^{-x^2} + 4x^2 e^{-x^2} $
$ f''(x) = e^{-x^2} (4x^2 - 2) $

Приравняем вторую производную к нулю и найдем критические точки.

$ e^{-x^2} (4x^2 - 2) = 0 $

$ 4x^2 - 2 = 0 $

$ 4x^2 = 2 $

$ x^2 = \frac{1}{2} $

$ x = \pm \frac{1}{\sqrt{2}} $

Определим знаки второй производной в интервалах между критическими точками.

 Для $ x < -\frac{1}{\sqrt{2}} $: $ f''(x) = e^{-x^2} (4x^2 - 2) < 0 $  
 Для $ -\frac{1}{\sqrt{2}} < x < \frac{1}{\sqrt{2}} $: $ f''(x) = e^{-x^2} (4x^2 - 2) > 0 $  
 Для $ x > \frac{1}{\sqrt{2}} $: $ f''(x) = e^{-x^2} (4x^2 - 2) < 0 $  

Определим интервалы выпуклости и точки перегиба.

 Функция выпукла вверх на интервалах $ (-\infty, -\frac{1}{\sqrt{2}}) $ и $ (\frac{1}{\sqrt{2}}, \infty) $.  
 Функция выпукла вниз на интервале $ (-\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}) $.  
 Точки перегиба: $ x = \pm \frac{1}{\sqrt{2}} $.

## 4. Найти асимптоты графиков функций:

1) $ y = \frac{3x}{x+2} $

Найдем вертикальные асимптоты.

Вертикальные асимптоты могут возникнуть там, где знаменатель функции обращается в ноль.

$ x + 2 = 0 $

$ x = -2 $

Таким образом, вертикальная асимптота: $ x = -2 $.

Найдем горизонтальные асимптоты.

Для нахождения горизонтальных асимптот, рассмотрим поведение функции при $ x \to \pm \infty $.

$ y = \frac{3x}{x+2} = \frac{3x}{x(1 + \frac{2}{x})} = \frac{3}{1 + \frac{2}{x}} $

При $ x \to \pm \infty $:

$ y \to \frac{3}{1 + 0} = 3 $

Таким образом, горизонтальная асимптота: $ y = 3 $.

2) $ y = e^{-\frac{1}{x}} $

Найдем вертикальные асимптоты.

Вертикальные асимптоты могут возникнуть там, где выражение под экспонентой обращается в бесконечность.

$ -\frac{1}{x} \to \infty \quad \text{при} \quad x \to 0^+ $
$ -\frac{1}{x} \to -\infty \quad \text{при} \quad x \to 0^- $

Таким образом, вертикальная асимптота: $ x = 0 $.

Найдем горизонтальные асимптоты.

Для нахождения горизонтальных асимптот, рассмотрим поведение функции при $ x \to \pm \infty $.

При $ x \to \infty $:

$ e^{-\frac{1}{x}} \to e^0 = 1 $

При $ x \to -\infty $:

$ e^{-\frac{1}{x}} \to e^0 = 1 $

Таким образом, горизонтальная асимптота: $ y = 1 $.

График функции $ y = x - \ln x $ убывает при $ 0 < x < 1 $, возрастает при $ x > 1 $. Он имеет минимум в точке $ (1, 1) $ и вертикальную асимптоту $ x = 0 $.

_Построим графикb функций._

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

fig, axs = plt.subplots(2, 2, figsize=(12, 12))

## y = ln(1 - x^2)
x1 = np.linspace(-0.99, 0.99, 400)
y1 = np.log(1 - x1**2)

def f1(x):
    return np.log(1 - x**2)


axs[0, 0].plot(x1, y1, label='$y = \\ln(1 - x^2)$')
axs[0, 0].set_title('График функции $y = \\ln(1 - x^2)$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].axvline(x=-1, color='red', linestyle='dashed', label='Асимптота x = -1')
axs[0, 0].axvline(x=1, color='red', linestyle='dashdot', label='Асимптота x = 1')
axs[0, 0].plot(0, 0, 'go', label='Критическая точка (0, 0)')

roots = [-1/np.sqrt(3), 1/np.sqrt(3)]
for root in roots:
    axs[0, 0].plot(root, f1(root), 'brown', marker='o', label=f'Точка перегиба ({root:.2f}, {f1(root):.2f})')

axs[0, 0].axhline(0, color='black', linewidth=0.5)
axs[0, 0].axvline(0, color='black', linewidth=0.5)
axs[0, 0].legend()

## y = x^2 / (1 - x^2)
x2 = np.linspace(-1.6, 1.6, 400)
y2 = x2**2 / (1 - x2**2)

axs[0, 1].plot(x2, y2, label='$y = \\frac{x^2}{1 - x^2}$')
axs[0, 1].set_title('График функции $y = \\frac{x^2}{1 - x^2}$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].axvline(x=-1, color='red', linestyle='dashed', label='Асимптота x = -1')
axs[0, 1].axvline(x=1, color='red', linestyle='dashdot', label='Асимптота x = 1')
axs[0, 1].axhline(y=-1, color='red', linestyle='dotted', label='Асимптота y = -1')
axs[0, 1].plot(0, 0, 'go', label='Критическая точка (0, 0)')
axs[0, 1].axhline(0, color='black', linewidth=0.5)
axs[0, 1].axvline(0, color='black', linewidth=0.5)
axs[0, 1].legend()

## y = x^2 * e^(-x)
x3 = np.linspace(-1, 6, 400)
y3 = x3**2 * np.exp(-x3)

def f3(x):
    return x**2 * np.exp(-x)

def f3_prime(x):
    return (2*x - x**2) * np.exp(-x)

def f3_double_prime(x):
    return (2 - 4*x + x**2) * np.exp(-x)

axs[1, 0].plot(x3, y3, label='$y = x^2 \\cdot e^{-x}$')
axs[1, 0].set_title('График функции $y = x^2 \\cdot e^{-x}$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].axhline(y=0, color='red', linestyle='dashdot', label='Асимптота y = 0')
axs[1, 0].plot(0, 0, 'go', label='Критическая точка (0, 0)')
axs[1, 0].plot(2, 4*np.exp(-2), 'go', label='Критическая точка (2, 4e^-2)')

roots = fsolve(f3_double_prime, [1, 3])
for root in roots:
    if -1 < root < 6:
        axs[1, 0].plot(root, f3(root), 'brown', marker='o', label=f'Точка перегиба ({root:.2f}, {f3(root):.2f})')

axs[1, 0].axhline(0, color='black', linewidth=0.5)
axs[1, 0].axvline(0, color='black', linewidth=0.5)
axs[1, 0].legend()

## y = x - ln(x)
x4 = np.linspace(0.01, 5, 400)
y4 = x4 - np.log(x4)

axs[1, 1].plot(x4, y4, label='$y = x - \\ln(x)$')
axs[1, 1].set_title('График функции $y = x - \\ln(x)$')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].axvline(x=0, color='red', linestyle='dashdot', label='Асимптота x = 0')
axs[1, 1].plot(1, 1, 'go', label='Критическая точка (1, 1)')
axs[1, 1].axhline(0, color='black', linewidth=0.5)
axs[1, 1].axvline(0, color='black', linewidth=0.5)
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
